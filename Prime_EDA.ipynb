```python
## importing libraries of python for analysis

import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Loading the file into notebook

df = pd.read_excel("E:\Tableau\projects\prime_cleaned.xlsx")
df.head()
```

```python
## Droping the extra indexed column

df = df.drop('Unnamed: 0',axis=1)
df.head()
```

```python
## Checking Number of columns and rows

df.shape
```

```python
## Checking the Datatypes

df.dtypes
```

```python
## Checking for the Null Values

df.isnull().sum()
```

```python
## Checking for Duplicates

df.duplicated().sum()
```

```python
## Modifying the rows in column genre

df['First_Word'] = df['genre'].str.split(',').str[0]
df.head()
```

```python
## Dropping the unwanted column

df = df.drop('genre',axis =1)
df.head()
```

```python
## Renaming the existing column

df.rename(columns={'First_Word': 'Genre'}, inplace=True)
df.head()
```

```python
## counting the number of Movies/TV shows released over the years

year_wise_movies = df['release_year'].value_counts().sort_index(ascending=True).tail(20)
year_wise_movies
```

```python
## Plotting the yearwise Movie/TV shows release

sns.set(rc={'figure.figsize':(15,5)})

ax = sns.countplot(x='release_year',data =df,order=year_wise_movies.index,hue='release_year')
plt.ylabel('Number of movies / tv shows')
plt.xlabel('Release year')
plt.title('YearWise Movie Release')

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.show()
```

```python
## counting the Top 10 Movies/TV shows based on Genre

genre_wise = df['Genre'].value_counts().sort_values(ascending=False).head(10)
genre_wise
```

```python
## Plotting the Top 10 Movie/TV shows based on Genre

ax = sns.countplot(x='Genre',data = df,order=genre_wise.index,hue='Genre')
plt.ylabel('Number of movies / tv shows',fontsize=13)
plt.title('Top 10 Most Watched Genres',fontsize=15)
plt.xlabel('Genre',fontsize=13)


for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
## counting the least number of Movies/TV shows based on Genre

genre_wise_least = df['Genre'].value_counts().sort_values(ascending=False).tail(5)
genre_wise_least
```

```python
## Plotting the Least Movie/TV shows based on Genre

ax = sns.countplot(x='Genre',data = df,order=genre_wise_least.index,hue='Genre')
plt.ylabel('Number of movies / tv shows',fontsize=13)
plt.title('Least watched Genres',fontsize=15)
plt.xlabel('Genre',fontsize=13)

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
## counting the Top 10 Movies/TV shows based on Ratings

rating_wise = df['rating'].value_counts().sort_values(ascending=False).head(10)
rating_wise
```

```python
## Plotting the Top 10 Movie/TV shows based on Ratings

ax = sns.countplot(x='rating',data=df,order=rating_wise.index,hue='rating',legend=False)

plt.ylabel('Number of movies / tv shows')
plt.title('No.of Movies based on rating')

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.show()
```

```python
# Plotting the Genre Distribution in percentage in pie chart using Matplotlib

plt.figure(figsize=(6, 6))
plt.pie(genre_wise,labels=genre_wise.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'),textprops={'fontsize':10})
plt.title('Genre Distribution using pie chart')
plt.show()
```

```python
## Counting the Movies/TV Shows by grouping release year and Genre

year_wise_genre = df.groupby(['release_year', 'Genre']).size().reset_index(name='count').sort_values(by='count',ascending=True).tail(20)
year_wise_genre
```

```python
## Plotting the Movies/TV Shows by grouping release year and Genre in LineChart

plt.figure(figsize=(6, 4))
sns.lineplot(x='release_year', y='count', hue='Genre', data=year_wise_genre)

plt.ylabel('Number of Movies/TV Shows')
plt.title('Genre Distribution Over last 5 years')
plt.legend(fontsize=9,loc='upper left')
plt.show()
```

```python
## Checking the distribution by Type 

type_wise = df['type'].value_counts()
type_wise
```

```python
## Plotting the type in pie chart

plt.figure(figsize=(4,4))
plt.pie(type_wise,labels=type_wise.index,autopct='%1.1f%%',startangle=90,colors=sns.color_palette('pastel'))
plt.title('Percentage of Movies/TV Shows')
plt.show()
```
